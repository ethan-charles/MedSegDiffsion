```python
import nibabel as nib
import matplotlib.pylab as plt
import numpy as np
```

```python
path = r'D:\umich_project_data\brain_dataset\MICCAI_BraTS2020_TrainingData\BraTS20_Training_009\BraTS20_Training_009_seg.nii'
```

```python
img = nib.load(path).get_fdata()
img.shape
```

```python
print(f"The .nii files are stored in memory as numpy's: {type(img)}.")
```

```python
plt.style.use('default')
fig, axes = plt.subplots(3, 3, figsize=(12,12))
for i, ax in enumerate(axes.reshape(-1)):
    ax.imshow(img[:,:,41 + i])
plt.show()
```

```python
# Use the default plotting style
plt.style.use('default')

# Create a single subplot
fig, ax = plt.subplots(figsize=(4, 4))  # Adjust the figure size as needed

# Display the 42nd slice of the image
ax.imshow(img[:, :, 41], cmap='gray')  # Use cmap='gray' for black and white

# Show the plot
plt.show()
```
